```python
import pandas as pd
```

```python
dataset = pd.read_csv("House_Size_vs_Price.csv")
```

```python
housesize = dataset[["House Size (sq ft)"]]
housesize
```

```python
price = dataset[["Price ($1000s)"]]
price
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split (housesize,price,test_size = 0.20,random_state =0)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor = LinearRegression()
regressor.fit(x_train,y_train)
```

```python
weight = regressor.coef_
weight
```

```python
bais = regressor.intercept_
bais
```

```python
y_pred = regressor.predict(x_test)
y_pred
```

```python
from sklearn.metrics import r2_score
r_score = r2_score(y_test,y_pred)
r_score
```

```python
import pickle
mymodel1 = "Finalized_Model_Linear.sav"
```

```python
pickle.dump(regressor,open (mymodel1,'wb'))
```

```python
loaded_model = pickle.load (open ("Finalized_Model_Linear.sav",'rb'))
loaded_model
```
