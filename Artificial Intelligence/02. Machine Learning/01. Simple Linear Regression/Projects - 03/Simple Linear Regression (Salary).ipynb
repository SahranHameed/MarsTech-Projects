```python
#Tablets,Contants Using LIB.
import pandas as pd #Calling Pandas.
```

```python
#var=clpanda .class fun cl All Data Stored In Variable 
dataset = pd.read_csv("Salary_Data.csv") 
```

```python
# Split Data Divide Y_Experience & store variable
independent = dataset[["YearsExperience"]] 
```

```python
# Split Data Divide Salary & store variable
dependent = dataset[["Salary"]]
```

```python
# Model Creation
```

```python
       #Only One Class Using         #Import Split Function 
from sklearn.model_selection import train_test_split
```

```python
#split Data Stored in x & y Variable                                           %testing     Blance % Training
x_train,x_test,  y_train,y_test = train_test_split(independent, dependent, test_size=0.30,random_state=0)
```

```python
#Using Linear Regression function
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train,y_train) # Learning Processing
```

```python
             # .class -> Slope=(M or W)
weight = regressor.coef_
weight # Slope
```

```python
       # .class    Bias=(C or B)
bias = regressor.intercept_
bias
```

```python
y_pred = regressor.predict(x_test)
y_pred
```

```python
y_test
```

```python
# Y testing function         Compair function(y_test,y_pred)
from sklearn.metrics import r2_score
r_score = r2_score(y_test,y_pred)
r_score
```
